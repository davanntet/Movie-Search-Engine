# clean data

```python

# read the dataset and store it in a dataframe
df = pd.read_csv("datasets/netflix_titles.csv")
df 
```

```python
# show the columns name of the dataframe
columns = df.columns
columns
```

```python
# create new variable to store the columns that we need
datas = df[['title','cast', 'release_year', 'rating', 'description','duration']]
```

```python
# show the first 5 rows of the dataframe
datas.head()
```

```python
#The selected code is performing data preprocessing on a pandas DataFrame named `datas`. 
#Specifically, it's handling missing values in the 'cast' and 'duration' columns.
datas['cast'] = datas['cast'].fillna('empty')
datas['duration'] = datas['duration'].fillna(datas['duration'].mode().values[0])
```

```python
# summarize the number of missing values in each column of the dataframe
datas.isnull().sum()
```

```python
# handle missing values in DataFrame by replacing them with the most common value in the column.
datas['rating'] = datas['rating'].fillna(str(datas['rating'].mode().values[0]))
```

```python
datas.isnull().sum()
```

```python
# show the first 5 rows of the dataframe
datas.head()
```

```python
# rename a column in DataFrame. 
#After this line is executed, all references to the 'cast' column should be replaced with 'actors'.
datas.rename(columns={'cast':'actors'}, inplace=True)
```

```python
# this line of code is a way to save a DataFrame to a CSV file. 
# After this line is executed, a new file named "netflix_titles_cleaned.csv" will be created in the "datasets" directory, and it will contain the data from the `datas` DataFrame.
datas.to_csv("datasets/netflix_titles_cleaned.csv", index=False)
```

```python
datas.isnull().sum()
```

# prepare data for training

```python
import pandas as pd
```

```python
# read the dataset and store it in a dataframe, this dataset is the dataset that we have cleaned above and saved it as a file csv
df = pd.read_csv("datasets/netflix_titles_cleaned.csv")
df.head()
```

```python
# below code, will convert some columns of dataframe to one sentence and store it in a new dataframe because models perform better for data in sentence format.
story_cols = df.columns
story_cols
storys = pd.DataFrame()
init_title = story_cols[0]
storys[init_title] = df[init_title].apply(lambda x: f"{init_title}: {x};")
story_cols = story_cols[1::]
for col in story_cols:
    storys[init_title] += df[col].apply(lambda x: f"{col}: {x};")
```

```python
# show the first 5 rows of the dataframe
storys.head()
```

```python
# save the dataframe as a csv file
storys.to_csv("datasets/storys.csv", index=False)
```

# convert data to sqlite

```python
import sqlite3
import numpy as np
```

```python
# read data from dataset that we have cleaned and saved it as a file csv
datas = pd.read_csv("datasets/netflix_titles_cleaned.csv")
```

```python
# create a new columns named 'ranking' and fill it with zeros
datas['ranking'] = np.zeros(datas.shape[0])
```

```python
datas.head()
```

```python
# create a connection to the database
conn = sqlite3.connect("datasets/netflix_titles.db")
```

```python
# save the dataframe as a table in the database
datas.to_sql("netflix_titles", conn, if_exists="replace")
```

```python
# test to query data from the database
conn.execute("select * from netflix_titles").fetchall()
```

```python
# close the connection to the database
conn.close()
```
